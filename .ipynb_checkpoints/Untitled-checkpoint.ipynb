# 1. Karakteristik Dataset Bike Sharing Dataset

Dataset Bike Sharing Dataset terkait dengan layanan penyewaan sepeda dan mencakup informasi harian serta per jam tentang pengguna yang menyewa sepeda. Terdapat dua file:

- **day.csv**: Informasi harian terkait penggunaan sepeda.
- **hour.csv**: Informasi penggunaan sepeda per jam.

**Fitur-Fitur Penting dalam Dataset:**

- **instant**: ID unik untuk setiap baris data.
- **dteday**: Tanggal kejadian.
- **season**: Musim dalam setahun (1: Winter, 2: Spring, 3: Summer, 4: Fall).
- **yr**: Tahun (0: 2011, 1: 2012).
- **mnth**: Bulan (1 sampai 12).
- **holiday**: Apakah hari tersebut adalah hari libur atau tidak (0: No, 1: Yes).
- **weekday**: Hari dalam seminggu (0: Minggu, 1: Senin, dst.).
- **workingday**: Apakah hari tersebut adalah hari kerja (0: Tidak, 1: Ya).
- **weathersit**: Kondisi cuaca (1: Clear, 2: Mist, 3: Light Snow/Rain, 4: Heavy Rain).
- **temp**: Suhu normal (dalam skala 0 sampai 1).
- **atemp**: Suhu yang dirasakan (dalam skala 0 sampai 1).
- **hum**: Kelembapan (dalam skala 0 sampai 1).
- **windspeed**: Kecepatan angin (dalam skala 0 sampai 1).
- **casual**: Jumlah pengguna casual (non-member).
- **registered**: Jumlah pengguna terdaftar (member).
- **cnt**: Total jumlah sepeda yang digunakan.

Dari dataset ini, saya ingin menganalisis beberapa hal, seperti:

- Tren musiman atau harian dalam penggunaan sepeda.
- Pengaruh cuaca atau hari kerja/libur terhadap penggunaan sepeda.
- Distribusi penggunaan sepeda antara pengguna casual dan registered.

### Descriptive Statistics
Descriptive statistics berguna untuk menganalisis karakteristik dasar dari data tersebut. Dan ini termasuk rata-rata (mean), median, mode, distribusi, dan lain-lain untuk setiap variabel penting.

#### Mengecek gambaran tentang distribusi data numerik seperti rata-rata, standar deviasi, nilai min/max, dll. pada file dataset day.csv

```python
df_day.describe()
```

#### Mengecek statistik deskriptif penggunaan sepeda berdasarkan musim

```python
df_day.groupby('season')['cnt'].describe()
```

#### Mengecek statistik deskriptif penggunaan sepeda berdasarkan cuaca

```python
df_day.groupby('weathersit')['cnt'].describe()
```

#### Mengecek statistik penggunaan sepeda berdasarkan hari kerja dan hari libur
